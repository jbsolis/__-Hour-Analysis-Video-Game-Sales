```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv("vgsales.csv")
```

```python
data.head()
```

```python
# Remove all games after 2016
data = data[data.Year < 2017.]
```

```python
sns.set(rc={"figure.figsize": (10, 10)})
sns.countplot(y = "Platform", data = data)
```
